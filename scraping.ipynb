```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup 
```

```python
url = "https://www.hubertiming.com/results/2017GPTR10K"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html , "lxml")
```

```python
title= soup.title
print(title)
```

```python
text = soup.get_text()
print(text[:10])
```

```python
soup.find_all('a')  #<a> hyperlink
```

```python
all_links = soup.find_all('a')
for link in all_links:
    print(link.get("href"))
```

```python
rows = soup.find_all('tr')

print(rows[:2])
```

```python
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)  
```

```python
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df.head(2)
```

```python
df1 = df[0].str.split(',', expand = True)
df1.head(2)
```

```python
df1[0] = df1[0].str.split('[')
df1.head(2)
```

```python
col_labels = soup.find_all('th')
all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "lxml").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
df2 = pd.DataFrame(all_header)
df2.head()
```

```python
df3 = df2[0].str.split(',', expand=True)
df3.head()
```

```python
frames = [df3, df1]

df4 = pd.concat(frames)
df4.head()
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
df5.info()
df5.shape
```

```python
df6 = df5.dropna(axis=0, how='any')   #dropped all null values
```

```python
df7 = df6.drop(df6.index[0])
df7.head()      #dropping null table headers
```

```python
df7.head()
```

```python
df7.rename(columns={'[Place': 'Place'},inplace=True)
df7.rename(columns={' Team]': 'Team'},inplace=True)
df7.head()
```

```python
df7['Team'] = df7['Team'].str.strip(']')
df7.head()
```

```python
df7.to_excel(r'./output/marathon.xlsx',index = False)
```
